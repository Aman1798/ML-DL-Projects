We will choose Classification feature as Petal Length to distinguish among the species

```python
iris_setosa = iris["species"]=="setosa"
new_iris_setosa = iris[iris_setosa]

iris_versicolor = iris["species"]=="versicolor"
new_iris_versicolor = iris[iris_versicolor]

iris_virginica = iris["species"]=="virginica"
new_iris_virginica = iris[iris_virginica]

```

```python
#CDF & PDF of petal_length wrt setosa
counts, bin_edges = np.histogram(new_iris_setosa["petal_length"], bins=10, density=True)


pdf = counts / (sum(counts)) # compute PDF
print(f'pdf = {pdf}')
print(f'bin_edges = {bin_edges}')


cdf = np.cumsum(pdf) # compute CDF

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend()
plt.gca().legend(('pdf', 'cdf'))
plt.show()
```

```python
#CDF & PDF of petal_length wrt versicolor
counts, bin_edges = np.histogram(new_iris_versicolor["petal_length"], bins=10, density=True)


pdf = counts / (sum(counts)) # compute PDF
print(f'pdf = {pdf}')
print(f'bin_edges = {bin_edges}')


cdf = np.cumsum(pdf) # compute CDF

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend()
plt.gca().legend(('pdf', 'cdf'))
plt.show()
```

```python
#CDF & PDF of petal_length wrt verginica
counts, bin_edges = np.histogram(new_iris_virginica["petal_length"], bins=10, density=True)


pdf = counts / (sum(counts)) # compute PDF
print(f'pdf = {pdf}')
print(f'bin_edges = {bin_edges}')


cdf = np.cumsum(pdf) # compute CDF

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend()
plt.gca().legend(('pdf', 'cdf'))
plt.show()
```

```python
#Box Plot
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.set_style("whitegrid")
sns.boxplot(x="species", y="petal_length", data=iris, orient='v',ax=axes[0, 1])
sns.boxplot(x="species", y="petal_width", data=iris, orient='v',ax=axes[0, 0])
sns.boxplot(x="species", y="sepal_length", data=iris,orient='v', ax=axes[1, 0])
sns.boxplot(x="species", y="sepal_width", data=iris,orient='v', ax=axes[1, 1])
plt.show()
```

-Setosa exhibits smaller features and a narrower distribution.
-Versicolor displays a moderate level of distribution and features of average size.
-Virginica showcases a high degree of distribution with a substantial number of values and large features.
